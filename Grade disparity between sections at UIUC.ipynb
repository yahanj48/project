## Introduction
The dataset comes from "wadefagen's Useful Datasets". This dataset contains a record of the grade distribution for each course taught by different professors. It starts from Spring 2010 to Summer 2020. There are 53933 instances in our data and 21 variables. The 'A+' to 'W' columns contain the information of how many students get which score. Here below shows the detailed information of other columns:

-`Year`: The scheduled year of the course.

-`Term`: The scheduled term of the course.

-`YearTerm`: The scheduled year and term of the course.

-`Subject`: The department abbreviation the course belongs to.

-`Number`: The course number

-`Course Title`: The name of the course.

-`Primary Instructor`: The professor who teaches the course.
